```python
# Prepare input (X) and output (y)
X = []
y = []
sec = 1
for reading in data:
    devices = reading["devices"]
    sample = []
    total_energy = 0
    for device in devices:
        energy_wh = device["energy_wh"]
        count = device["count"]
        
        # Append [energy, count] as input
        sample.append([energy_wh, count, sec])
```

```python
# Compute correct total
total_energy += energy_wh * count

X.append(sample)
y.append(total_energy*sec)
sec += 1
# Convert to numpy arrays
X = np.array(X)  # shape: (n_samples, 9, 2)
y = np.array(y)  # shape: (n_samples,)
```
